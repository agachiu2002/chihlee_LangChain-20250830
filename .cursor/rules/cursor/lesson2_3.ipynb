## 1. while è¿´åœˆåŸºæœ¬ç”¨æ³•
while è¿´åœˆæœƒåœ¨æ¢ä»¶ç‚º True æ™‚æŒçºŒåŸ·è¡Œï¼Œç›´åˆ°æ¢ä»¶è®Šç‚º False


```python
# ç¯„ä¾‹ 1: åŸºæœ¬çš„ while è¿´åœˆ
# å¾ 1 æ•¸åˆ° 5
count = 1
while count <= 5:
    print(f"è¨ˆæ•¸: {count}")
    count += 1  # é‡è¦ï¼šè¨˜å¾—è¦æ”¹è®Šæ¢ä»¶ï¼Œå¦å‰‡æœƒç„¡é™è¿´åœˆ
```

## 2. break èªå¥
break æœƒç«‹å³è·³å‡ºè¿´åœˆï¼Œä¸ç®¡æ¢ä»¶æ˜¯å¦é‚„ç‚º True


```python
# ç¯„ä¾‹ 2: ä½¿ç”¨ break è·³å‡ºè¿´åœˆ
# æ‰¾åˆ°ç¬¬ä¸€å€‹å¤§æ–¼ 10 çš„æ•¸å­—å°±åœæ­¢
numbers = [1, 3, 7, 12, 15, 20]
i = 0

while i < len(numbers):
    print(f"æª¢æŸ¥æ•¸å­—: {numbers[i]}")
    if numbers[i] > 10:
        print(f"æ‰¾åˆ°å¤§æ–¼ 10 çš„æ•¸å­—: {numbers[i]}ï¼Œåœæ­¢æœå°‹")
        break  # è·³å‡ºè¿´åœˆ
    i += 1

print("è¿´åœˆçµæŸ")
```

## 3. continue èªå¥
continue æœƒè·³éç•¶å‰è¿´åœˆçš„å‰©é¤˜éƒ¨åˆ†ï¼Œç›´æ¥é€²å…¥ä¸‹ä¸€æ¬¡è¿´åœˆ


```python
# ç¯„ä¾‹ 3: ä½¿ç”¨ continue è·³éæŸäº›æ•¸å­—
# åªå°å‡ºå¥‡æ•¸ï¼Œè·³éå¶æ•¸
count = 1
while count <= 10:
    if count % 2 == 0:  # å¦‚æœæ˜¯å¶æ•¸
        count += 1
        continue  # è·³éå¾Œé¢çš„ç¨‹å¼ç¢¼ï¼Œç›´æ¥é€²å…¥ä¸‹ä¸€æ¬¡è¿´åœˆ
    print(f"å¥‡æ•¸: {count}")
    count += 1
```

```python
# ç¯„ä¾‹ 5: è¨ˆç®—éšä¹˜ (n!)
# éšä¹˜ï¼šn! = n Ã— (n-1) Ã— (n-2) Ã— ... Ã— 1
# ä¾‹å¦‚ï¼š5! = 5 Ã— 4 Ã— 3 Ã— 2 Ã— 1 = 120

def factorial(n):
    """è¨ˆç®— n çš„éšä¹˜"""
    if n < 0:
        return "éšä¹˜åªèƒ½è¨ˆç®—éè² æ•´æ•¸"
    
    result = 1
    i = 1
    
    while i <= n:
        result *= i
        i += 1
    
    return result

# æ¸¬è©¦éšä¹˜å‡½æ•¸
print("éšä¹˜è¨ˆç®—ç¯„ä¾‹ï¼š")
for num in range(1, 6):
    print(f"{num}! = {factorial(num)}")
```

```python
# ç¯„ä¾‹ 4: çŒœæ•¸å­—éŠæˆ²
import random

# ç”¢ç”Ÿ 1 åˆ° 100 çš„éš¨æ©Ÿæ•¸å­—
secret_number = random.randint(1, 100)
attempts = 0
max_attempts = 7

print("æ­¡è¿ä¾†åˆ°çŒœæ•¸å­—éŠæˆ²ï¼")
print("æˆ‘å·²ç¶“æƒ³å¥½ä¸€å€‹ 1 åˆ° 100 ä¹‹é–“çš„æ•¸å­—")
print(f"ä½ æœ‰ {max_attempts} æ¬¡æ©ŸæœƒçŒœä¸­å®ƒï¼")

while attempts < max_attempts:
    try:
        guess = int(input(f"ç¬¬ {attempts + 1} æ¬¡çŒœæ¸¬ï¼Œè«‹è¼¸å…¥ä½ çš„çŒœæ¸¬ (1-100): "))
        
        # æª¢æŸ¥è¼¸å…¥æ˜¯å¦åœ¨æœ‰æ•ˆç¯„åœå…§
        if guess < 1 or guess > 100:
            print("è«‹è¼¸å…¥ 1 åˆ° 100 ä¹‹é–“çš„æ•¸å­—ï¼")
            continue  # è·³éé€™æ¬¡çŒœæ¸¬ï¼Œä¸è¨ˆå…¥æ¬¡æ•¸
        
        attempts += 1
        
        if guess == secret_number:
            print(f"ğŸ‰ æ­å–œä½ ï¼ä½ çŒœå°äº†ï¼ç­”æ¡ˆæ˜¯ {secret_number}")
            print(f"ä½ ç¸½å…±ç”¨äº† {attempts} æ¬¡çŒœä¸­ï¼")
            break  # çŒœå°äº†ï¼Œè·³å‡ºè¿´åœˆ
        elif guess < secret_number:
            print("ğŸ“ˆ å¤ªå°äº†ï¼å†è©¦ä¸€æ¬¡")
        else:
            print("ğŸ“‰ å¤ªå¤§äº†ï¼å†è©¦ä¸€æ¬¡")
            
        print(f"é‚„å‰©ä¸‹ {max_attempts - attempts} æ¬¡æ©Ÿæœƒ")
        
    except ValueError:
        print("è«‹è¼¸å…¥æœ‰æ•ˆçš„æ•¸å­—ï¼")
        continue  # è¼¸å…¥éŒ¯èª¤ï¼Œä¸è¨ˆå…¥æ¬¡æ•¸

# æª¢æŸ¥æ˜¯å¦ç”¨å®Œäº†æ‰€æœ‰æ©Ÿæœƒ
if attempts >= max_attempts and guess != secret_number:
    print(f"ğŸ˜¢ éŠæˆ²çµæŸï¼ç­”æ¡ˆæ˜¯ {secret_number}")
    print("ä¸‹æ¬¡å†ä¾†æŒ‘æˆ°å§ï¼")
```
