```python
# 數學運算範例
first_number = 7
second_number = 3

addition_result = first_number + second_number
subtraction_result = first_number - second_number
multiplication_result = first_number * second_number
division_result = first_number / second_number
remainder_result = first_number % second_number
exponent_result = first_number ** second_number

print(f"加法 {first_number} + {second_number} = {addition_result}")
print(f"減法 {first_number} - {second_number} = {subtraction_result}")
print(f"乘法 {first_number} * {second_number} = {multiplication_result}")
print(f"除法 {first_number} / {second_number} = {division_result:.2f}")
print(f"取餘數 {first_number} % {second_number} = {remainder_result}")
print(f"次方 {first_number} ** {second_number} = {exponent_result}")

```

```python
# 有一個直角三角形，已知兩邊長，求第三邊長
# 已知兩邊長，求第三邊長

import math

adjacent_side = 3  # 鄰邊長度
opposite_side = 4  # 對邊長度

# 使用畢氏定理計算斜邊長度
hypotenuse_length = math.sqrt(adjacent_side**2 + opposite_side**2)

print(f"斜邊長度 = {hypotenuse_length:.2f}")
```

```python
# 更多數學運算範例
print("=== 進階數學運算範例 ===")

# 使用繁體中文變數名稱
第一個數字 = 10
第二個數字 = 4
第三個數字 = 2

print(f"第一個數字: {第一個數字}")
print(f"第二個數字: {第二個數字}")
print(f"第三個數字: {第三個數字}")
print()

# 複合運算
總和 = 第一個數字 + 第二個數字 + 第三個數字
print(f"三個數字相加: {第一個數字} + {第二個數字} + {第三個數字} = {總和}")

# 平均值計算
平均值 = 總和 / 3
print(f"平均值: {總和} ÷ 3 = {平均值:.2f}")

# 面積計算 (長方形)
長度 = 8
寬度 = 5
面積 = 長度 * 寬度
print(f"長方形面積: 長度 {長度} × 寬度 {寬度} = {面積}")

# 圓周率計算
半徑 = 6
圓周率 = 3.14159
圓面積 = 圓周率 * 半徑 ** 2
print(f"圓面積: π × {半徑}² = {圓面積:.2f}")
```

```python
# 開平方根運算範例
import math  # 匯入數學模組

print("=== 開平方根運算範例 ===")

# 方法一：使用 math.sqrt() 函數
數字1 = 16
平方根1 = math.sqrt(數字1)
print(f"方法一 - math.sqrt(): √{數字1} = {平方根1}")

數字2 = 25
平方根2 = math.sqrt(數字2)
print(f"方法一 - math.sqrt(): √{數字2} = {平方根2}")

數字3 = 9
平方根3 = math.sqrt(數字3)
print(f"方法一 - math.sqrt(): √{數字3} = {平方根3}")

print()

# 方法二：使用次方運算 (x ** 0.5)
print("=== 使用次方運算計算平方根 ===")
數字4 = 36
平方根4 = 數字4 ** 0.5
print(f"方法二 - 次方運算: {數字4} ** 0.5 = {平方根4}")

數字5 = 49
平方根5 = 數字5 ** 0.5
print(f"方法二 - 次方運算: {數字5} ** 0.5 = {平方根5}")

數字6 = 64
平方根6 = 數字6 ** 0.5
print(f"方法二 - 次方運算: {數字6} ** 0.5 = {平方根6}")

print()

# 小數的平方根
print("=== 小數的平方根 ===")
小數1 = 2.25
小數平方根1 = math.sqrt(小數1)
print(f"√{小數1} = {小數平方根1}")

小數2 = 6.25
小數平方根2 = math.sqrt(小數2)
print(f"√{小數2} = {小數平方根2}")

print()

# 實用範例：計算直角三角形斜邊長度
print("=== 實用範例：直角三角形斜邊長度 ===")
直角邊1 = 3
直角邊2 = 4
斜邊長度 = math.sqrt(直角邊1**2 + 直角邊2**2)
print(f"直角邊1: {直角邊1}")
print(f"直角邊2: {直角邊2}")
print(f"斜邊長度: √({直角邊1}² + {直角邊2}²) = √({直角邊1**2} + {直角邊2**2}) = {斜邊長度}")

print()

# 實用範例：計算圓的半徑（已知面積）
print("=== 實用範例：計算圓的半徑 ===")
圓面積 = 78.54
圓半徑 = math.sqrt(圓面積 / math.pi)
print(f"圓面積: {圓面積}")
print(f"圓半徑: √(面積 ÷ π) = √({圓面積} ÷ {math.pi:.2f}) = {圓半徑:.2f}")

print()

# 實用範例：計算距離（兩點間距離）
print("=== 實用範例：兩點間距離 ===")
點1x = 0
點1y = 0
點2x = 3
點2y = 4
距離 = math.sqrt((點2x - 點1x)**2 + (點2y - 點1y)**2)
print(f"點1: ({點1x}, {點1y})")
print(f"點2: ({點2x}, {點2y})")
print(f"距離: √(({點2x}-{點1x})² + ({點2y}-{點1y})²) = {距離}")
```

```python
# 運算優先順序範例
print("=== 運算優先順序範例 ===")

# 括號優先
括號運算 = (a + b) * 2
print(f"括號運算: ({a} + {b}) × 2 = {括號運算}")

# 次方優先
次方優先 = a + b ** 2
print(f"次方優先: {a} + {b}² = {次方優先}")

# 乘除優先於加減
乘除優先 = a + b * 2
print(f"乘除優先: {a} + {b} × 2 = {乘除優先}")

print()
print("=== 實用計算範例 ===")

# 溫度轉換 (攝氏轉華氏)
攝氏溫度 = 25
華氏溫度 = 攝氏溫度 * 9 / 5 + 32
print(f"溫度轉換: {攝氏溫度}°C = {華氏溫度}°F")

# 購物計算
商品單價 = 150
購買數量 = 3
折扣率 = 0.9  # 9折
小計 = 商品單價 * 購買數量
折扣後金額 = 小計 * 折扣率
print(f"購物計算:")
print(f"  商品單價: {商品單價} 元")
print(f"  購買數量: {購買數量} 個")
print(f"  小計: {小計} 元")
print(f"  折扣後: {折扣後金額} 元")
```
